```python
#Modified while loop

balance = 0
deposit = 20500

age = 22

rate = 13.5
real_rate = rate*0.01

retirement_value = 35000000
while balance < retirement_value:
    balance = (balance + deposit)*(1 + real_rate)
    age += 1
    
print("I will have {} at {} years old." .format(balance, age))
```
